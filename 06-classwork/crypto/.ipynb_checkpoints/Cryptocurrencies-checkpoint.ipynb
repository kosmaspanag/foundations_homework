* Year extraction
* pd.to_datetime
* replace commas, or thousands=','
* fill_between? - no there isn't enough movement

```python
import pandas as pd
```

```python
df = pd.read_csv("cryptocurrencies.csv", thousands=',', parse_dates=True)
df.head()
```

```python
df['date'] = pd.to_datetime(df.date, format="%d-%b-%y")
df.head()
```

```python
%matplotlib inline

df.plot(x='date', y='open')
```

```python
df['open'] = df.open.str.replace(",", "").astype(float)
df['volume'] = df.volume.str.replace(",", "").astype(float)
df['market_cap'] = df.market_cap.str.replace(",", "").astype(float)
df['high'] = df.high.str.replace(",", "").astype(float)
df['low'] = df.low.str.replace(",", "").astype(float)
df['close'] = df.close.str.replace(",", "").astype(float)
```

```python
import matplotlib.pyplot as plt

df = df[df.coin == 'BTC']
plt.fill_between(df.date, df.low, df.high)
```
