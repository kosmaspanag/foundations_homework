```python
import pandas as pd
import requests

%matplotlib inline
```

```python
response = requests.get('https://api.darksky.net/forecast/15074402c474e3dab67f7377e1f95519/37.8267,-122.4233')
data = response.json()
```

```python
data
```

```python
#This is the hourly forecast
data['hourly']['data']
```

```python
pd.DataFrame(data['hourly']['data'])
```

```python
df = pd.DataFrame(data['hourly']['data'])
df.head()
```

```python
df.shape
```

```python
df.apparentTemperature.max()
```

```python
#The five warmest temperature predictions over the next 48 hours
df.sort_values(by='temperature', ascending=False).head(5)
```
