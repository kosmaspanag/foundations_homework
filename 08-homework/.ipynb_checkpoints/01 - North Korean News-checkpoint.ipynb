```python
import requests
from bs4 import BeautifulSoup
```

```python
my_url = "http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf"
raw_html = requests.get(my_url).content
```

```python
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(type(soup_doc))
```

```python
print(soup_doc.prettify())
```

```python
soup_doc.title.string
```

```python
soup_doc.find_all('div')
```

```python
titles = soup_doc.find_all('a', class_='titlebet')
titles
```

```python
headilne = []
for title in titles:
    print(title.string)
```

```python
onclicks = soup_doc.find_all('a', class_='titlebet')
onclicks
```

```python
for item in onclicks:
    print(item['onclick'])
```

```python
items = soup_doc.find_all('a', class_='titlebet')

news = []
for item in items:
    korean_dict = {}
    korean_dict['title'] = item.string
    korean_dict['onclick'] = item['onclick'] 
    news.append(korean_dict)

news
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(news)
df.head()
```
