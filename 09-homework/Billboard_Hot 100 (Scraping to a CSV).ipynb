# Billboard: Hot 100 (Scraping to a CSV)

https://www.billboard.com/charts/hot-100

Scrape the fields below, and save as a CSV file.

* Rank
* Artist name
* Song name

This one is pretty simple.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
```

```python
ranks = doc.find_all(class_="chart-list-item__rank")
for rank in ranks:
    print(rank.text.strip())
```

```python
names = doc.find_all(class_="chart-list-item__artist")
for name in names:
    print(name.text.strip())
```

```python
song_names = doc.find_all(class_="chart-list-item__title-text")
for song_name in song_names:
    print(song_name.text.strip())
```

```python
rows = []
songs = doc.find_all(class_='chart-list-item')

for song in songs:
    row = {}
    try:
        row['rank'] = song.find(class_='chart-list-item__rank').text.strip()
        row['artist_name'] = song.find(class_='chart-list-item__artist').text.strip()
        row['song name'] = song.find(class_='chart-list-item__title-text').text.strip()
    
    except:
        pass
    rows.append(row)


df = pd.DataFrame(rows)
df
```

```python
df.to_csv('Billboard_Hot_100.csv', index=False)
```
